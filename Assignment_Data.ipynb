```python
# Data Cleaning Steps

# 1. Handling NaN values: Fill numerical NaNs with median and categorical NaNs with 'unknown'
Assignment_Data_df['age'].fillna(Assignment_Data_df['age'].median(), inplace=True)
Assignment_Data_df['job'].fillna('unknown', inplace=True)
Assignment_Data_df['balance'].fillna(Assignment_Data_df['balance'].median(), inplace=True)

# As 'education', 'contact', and 'poutcome' already have 'unknown' values, we consider them as a category.

# 2. Handling categorical data: Check if conversion to numerical is needed
# For this dataset, we'll keep them as is for now since we can analyze them without conversion.

# Quick overview after cleaning
Assignment_Data_df.isnull().sum()  # Checking for remaining NaN values

# Let's also describe the dataset to understand its numerical features better.
Assignment_Data_df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the visualizations
sns.set_style("whitegrid")

# Visualization 1: Histogram of Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(Assignment_Data_df['age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Visualization 2: Average Account Balance by Job Category
plt.figure(figsize=(14, 8))
avg_balance_by_job = Assignment_Data_df.groupby('job')['balance'].mean().sort_values()
sns.barplot(x=avg_balance_by_job.values, y=avg_balance_by_job.index)
plt.title('Average Account Balance by Job')
plt.xlabel('Average Account Balance')
plt.ylabel('Job')

plt.tight_layout()
plt.show()
```

What is the distribution of account balances across different jobs?


```python
# Visualization: Box Plot for Account Balance Distribution across Different Jobs

plt.figure(figsize=(16, 10))
sns.boxplot(x='job', y='balance', data=Assignment_Data_df)
plt.xticks(rotation=45)
plt.title('Distribution of Account Balances Across Different Jobs')
plt.xlabel('Job')
plt.ylabel('Account Balance')
plt.tight_layout()
plt.show()
```

How does marital status affect the loan status (having a personal loan)?

```python
# Visualization: Marital Status vs. Loan Status

plt.figure(figsize=(10, 6))
sns.countplot(x='marital', hue='loan', data=Assignment_Data_df)
plt.title('Marital Status vs. Loan Status (Having a Personal Loan)')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.legend(title='Loan Status', loc='upper right')
plt.tight_layout()
plt.show()
```

What is the relationship between education level and the success of the campaign (y column)?


```python
# Visualization: Education Level vs. Success of the Campaign

plt.figure(figsize=(10, 6))
sns.countplot(x='education', hue='y', data=Assignment_Data_df)
plt.title('Education Level vs. Success of the Campaign')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.legend(title='Campaign Outcome', labels=['No', 'Yes'], loc='upper right')
plt.tight_layout()
plt.show()
```

How does the age distribution look among individuals with different educational backgrounds? 

```python
# Visualization: Age Distribution among Different Educational Backgrounds

plt.figure(figsize=(14, 8))
sns.violinplot(x='education', y='age', data=Assignment_Data_df)
plt.title('Age Distribution Among Individuals with Different Educational Backgrounds')
plt.xlabel('Education Level')
plt.ylabel('Age')
plt.tight_layout()
plt.show()
```
